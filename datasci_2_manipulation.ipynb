Load Packages

```python
import pandas as pd
import time
```

Reading CSV file to obtain column names

```python
df = pd.DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/jward6301/datasci_2_manipulation/main/datasets/healthcare_data_cleaning.csv')
df
```

```python
df.columns
```

Data Cleaning with Pandas

```python
#identifying missing values and counting them
missing_values = df.isnull()
print(missing_values)
```

```python
#identifying and counting missing values
missing_count = df.isnull().sum()
print(missing_count)
```

```python
#removing duplicate rows
df.drop_duplicates(inplace=True)
df
```

```python
#removing missing values
df.dropna(inplace=True)
df
```

```python
#Replacing unwanted characters in column names such as spaces
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.replace(r'[^a-zA-z0-9]', '')
df.columns = df.columns.str.lower()
df
```

```python
#extra line to furhter clean up column names
df.columns = df.columns.str.strip()
df
```

Data Transformation

```python
#Creating new columns
df['payment_category'] = df['payment_method'].apply(lambda x: 'Self-Pay' if x in ['Card', 'Cash'] else 'Insurance')
df
```

```python
#Creating new columns
df['review_patient'] = df['patient_age'].apply(lambda x: 'Needs Review' if x in ['missing'] else '')
df
```

```python
#Creating new columns
df['monitor_heart_rate'] = df['average_heart_rate'].apply(lambda x:'High' if x >= '120' else 'Normal')
df
```

```python
#groupby
grouped = df.groupby('state_of_residence')
df.head()
```

```python
df.groupby(['state_of_residence']).describe()
```

```python
count_values = grouped['state_of_residence'].count()
print(count_values)
```

```python
#pivot table

df.pivot_table(index='gender', columns='patient_age', values='height_in_cm', aggfunc='sum')
```
